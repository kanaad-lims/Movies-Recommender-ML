```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
movies.shape
```

```python
credits.head()
credits.shape
```

```python
#Merging the 2 datasets on the basis of the 'title' attribute.

movies = movies.merge(credits, on='title')
movies.head(1)
```

```python
#Columns to keep:
#genres
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]

movies.head()
```

```python
movies.isnull().sum() #Checking null values.
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]

#['Action', 'Adventure', 'Fantasy', 'SciFi']

#We want to display the data from the scattered format into much cleaner format. We will do this using a helper function.
```

```python
import ast
#We have created the helper function convert(), but we cannot use it as the obj to be passed from the genres is a string.
#So we will use a function 'ast' to convert string to list.

#HELPER 1

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
#We will apply the helper function on the genres column. Then store the converted data back into the genres column.

movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#Now, we will do the same things for the columns- keywords, cast and crew.

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#For cast, we need only the top 3 actors in the movie.
#Lets make anothre helper function to address this issue.

# HELPER 2

import ast
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
# we will apply it on the cast column.

movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# From the crew column, we need only the director name.
import ast
def fetchDir(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetchDir)
```

```python
movies.head()
```

```python
#The overview of the movies is in string format. So we have to convert it into a list so as to merge with the other columns.

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# We need to remove the spaces between two words so as to make them into a single enitity.

movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
#Now we have to concatenate genres, overview, keywords, cast, crew into a single column - 'tags' 

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(3)
```

```python
#We now dont need the concatenated columns. so lets create a new df which will consits of only id, title and tags columns.

new_df = movies[['movie_id', 'title', 'tags']]

new_df.head()
```

```python
#Now, the tags column data is in List as we converted using the ast function. We will then join them at the spaces.
#.join method converts the list of strings, into a single string. Here it uses the " " as a seperator.
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
#Converting everything into lower case (Recommend to make everything into lower case).

new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())

new_df.head()
```

```python
#We will apply Stemming technique so that similar words like say 'action' and 'actions' are taken as the base word 'action'
#We will use a library called nltk.

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#Creating a helper function to help the stemming process.
# when we do ps.stem('loving') ---> the answer is 'love'

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)  #Converts back into string.
    
```

```python
#Applying the helper function on tags.

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#Text Vectorization using the Bag of words approach
#We will be creating vectors for each movie in the 5000 dim space, with the 5000 most occuring words (ignoring the stop word in english language).

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() # new_df is converted into a numpy array. And is stored in a variable vectors.
#Each movie is basically converted into a vector.
```

```python
#First movie as a vector
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
#As we are dealing with higher dimensionality, we wont be using Euclidean distance to find relativeness between 2 movies.
#Instead we will be using Cosine distance, basically the angle between 2 vectors.

from sklearn .metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)

similarity[0]
#The result represents the distance between the first movie with each of the 4806 movies that are there.
```

```python
#Now, in the main function, when we sort, then the sorting will happen but the indexes will be changed.
#So like say for first movie, the distance with itself will be 1. So when we sort, this index will be changed.
#So then, we cannot say that first movie is this much similar with first/second/third movie, since the indexes are now changed.
#So, we use enumerate and list to maintain the indexes and still sort the distances.

sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1])[1:6] # for 5 movies
```

```python
#Main Function building
#Fetch the index of any selected movie, so that we can sort the similarities array and display the top 5 similar movies.
#using new_df[new_df['title'] == 'Batman Begins'].index[0]

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1]) [1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
#Example working of the function.
recommend('Batman Begins')
```
